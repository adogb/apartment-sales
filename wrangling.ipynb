```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

# Gathering data

```python
url = 'https://www.boliga.dk/resultat?propertyType=3&zipCodes=2720'
r = requests.get(url)
content=r.content
soup=BeautifulSoup(content)
```

```python
tag_list=soup.find("app-housing-list-results").find("div").find("div").next_sibling.find_all("app-housing-list-item")
```

```python
data_list = []
for tag in tag_list:
    if tag.find("app-listing-information-hidden"):
        #print("skuffesag")
        continue
    address=tag.find("app-listing-information-lg").div.div.div.div.span.string
    #print(address)
    city=address.parent.next_sibling.string
    #print(city)
    price=tag.find("app-listing-information-lg").div.div.div.next_sibling.div.contents[1]
    #print(price)
    pricePerM2=price.parent.next_sibling.string
    #print(pricePerM2)
    daysPosted = tag.find("app-listing-information-lg").div.contents[3].div.div.contents[2].contents[3]
    daysPosted = str(daysPosted).extract(r'(\d+)')
    
    # appending data dictionnary to data_list. BeautifulSoup converts HTML non-breaking spaces (&nbsp) to \xa0 (unicode),
    # so we need to remove them.
    data_list.append({
        "address": str(address).replace(u'\xa0', u' '),
        "city": str(city).replace(u'\xa0', u' '),
        "price": str(price).replace(u'\xa0', u' '),
        "pricePerM2": str(pricePerM2).replace(u'\xa0', u' '),
        "datePosted": str(daysPosted).replace(u'\xa0', u' ')
    })
    
```
